```python
from transformers import DPRQuestionEncoder, DPRQuestionEncoderTokenizer, DPRContextEncoder, DPRContextEncoderTokenizer
import torch
```

```python
chkp = "crystina-z/mdpr-tied-msmarco-pyserini"
```

```python
model_question = DPRQuestionEncoder.from_pretrained(chkp)
model_context = DPRContextEncoder.from_pretrained(chkp)
tokenizer_context = DPRContextEncoderTokenizer.from_pretrained(chkp)
tokenizer_question = DPRQuestionEncoderTokenizer.from_pretrained(chkp)
```

```python
sim = torch.nn.CosineSimilarity(dim=1, eps=1e-6)
```

```python
def get_q_vec(sent, tokenizer, model):
    input_ids = tokenizer(sent, return_tensors="pt")["input_ids"]
    embeddings = model(input_ids).pooler_output
    return embeddings
```

```python
question = 'อยุธยามีกี่อำเภอ ?'
context_1 = 'ละครครั้งที่ 3'
context_2 = 'ปัจจุบันจังหวัดพระนครศรีอยุธยาประกอบด้วย 16 อำเภอ ได้แก่'
```

```python
emb_q = get_q_vec(question, tokenizer_question, model_question)
print(emb_q.shape)
emb_q
```

```python
emb_c1 = get_q_vec(context_1, tokenizer_context, model_context)
print(emb_c1.shape)
emb_c1
```

```python
emb_c2 = get_q_vec(context_2, tokenizer_context, model_context)
print(emb_c2.shape)
emb_c2
```

```python
sim(emb_q, emb_c1)
```

```python
sim(emb_q, emb_c2)
```
